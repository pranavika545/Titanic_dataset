```python
# Importing modules
import numpy as np
import pandas as pd
```

```python
# Importing dataset 
train_data = pd.read_csv("./train.csv")
```

```python
# Importing important columns after analysis 
train_data = train_data[["Survived","Pclass","Sex","Age"]]
```

```python
# Handle null value
train_data["Age"] = train_data["Age"].fillna(train_data["Age"].mean())
train_data = pd.get_dummies(train_data,columns=["Sex"])
```

```python
train_data
```

```python
# Collecting data for training
X = train_data[["Pclass","Sex_female","Sex_male","Age"]]
Y = train_data["Survived"]
```

```python
train_data.info()
```

```python
train_data.info()
```

```python
# Getting decision tree classifier
from sklearn.ensemble import RandomForestClassifier

tcf = RandomForestClassifier()
tcf.fit(X,Y)
```

```python
# Importing Test data
test_data = pd.read_csv("test.csv")
```

```python
test_data.info()
```

```python
test_data["Age"] = test_data["Age"].fillna(test_data["Age"].mean())
test_data = pd.get_dummies(test_data,columns=["Sex"])
```

```python
test_data
```

```python
# Transforming dataset
X_test = test_data[["Pclass","Sex_female","Sex_male","Age"]]
```

```python
# Creating data frame for passender Id and survived 
out_values = pd.DataFrame(data={"PassengerId":test_data["PassengerId"],
                        "Survived":tcf.predict(X_test)})
```

```python
# Creating sumbission file
out_values.to_csv("submission_3.csv",index=False)
```
